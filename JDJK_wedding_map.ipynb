```python
import matplotlib as mpl
%matplotlib inline
from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from geopy.geocoders import Nominatim
```

```python
location_list = ["Maui Hawaii", 
 "Kona Hawaii", 
 "Kauai Hawaii", 
 "Tokyo", 
 "Singapore", 
 "Anaheim, California",
"San Diego, California",
"Los Angeles, California",
"Napa, California",
"Yosemite, California",
"Monterey, California",
"Edinburgh, Scotland",
"London, England",
"New Zealand",
"Ashland, Oregon",
"South lake Tahoe"]
```

```python
lat_list = []
long_list = []

for i in range(len(location_list)):

    geolocator = Nominatim(user_agent='myapplication')
    location = geolocator.geocode(location_list[i])
    lat_list.append(location.latitude)
    long_list.append(location.longitude)
```

```python
data = pd.DataFrame({
'lat':lat_list,
'lon':long_list,
'name':location_list
})

# https://peak5390.wordpress.com/2012/12/08/matplotlib-basemap-tutorial-plotting-points-on-a-simple-map/
```

```python
data
```

```python
plt.rcParams["figure.figsize"] = (15,8)

# initialise the map
m=Basemap()

# Control the background color
m.drawmapboundary(fill_color='#A6CAE0', linewidth=0)
 
# Fill the continent
m.fillcontinents(color='grey', alpha=0.7, lake_color='grey')
 
# Draw the coastline
m.drawcoastlines(linewidth=0.1, color="white")
 
# to save if needed
#plt.savefig('PNG/#282_Custom_Basemap.png', dpi=110, bbox_inches='tight')
 
# Show
# plt.show()
# lons = [0, 10, -20, -20]
# lats = [0, -10, 40, -20]
x, y = m(data['lon'].values, data['lat'].values)
# x, y = m(tmp[0], tmp[1])

m.plot(x, y, marker="o", 
       linestyle='none', 
       markersize=15, 
       alpha=0.6, 
       c="teal", 
       markeredgecolor="black", 
       markeredgewidth=1)

plt.savefig('JDJK.jpg', bbox_inches='tight')
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (15,12)

# initialise the map
m=Basemap(llcrnrlon=-130, llcrnrlat=20,urcrnrlon=-110,urcrnrlat=50)

# Control the background color
m.drawmapboundary(fill_color='#A6CAE0', linewidth=0)
 
# Fill the continent
m.fillcontinents(color='grey', alpha=0.7, lake_color='grey')
 
# Draw the coastline
m.drawcoastlines(linewidth=0.1, color="white")
m.drawstates(linewidth=0.1, color="white")
m.drawcountries(linewidth=0.1, color="white")

 
# to save if needed
#plt.savefig('PNG/#282_Custom_Basemap.png', dpi=110, bbox_inches='tight')
 
# Show
# plt.show()
# lons = [0, 10, -20, -20]
# lats = [0, -10, 40, -20]
x, y = m(data['lon'].values, data['lat'].values)
# x, y = m(tmp[0], tmp[1])

m.plot(x, y, marker="o", 
       linestyle='none', 
       markersize=14, 
       alpha=0.6, 
       c="teal", 
       markeredgecolor="black", 
       markeredgewidth=1)

plt.savefig('JDJK_zoom.jpg', bbox_inches='tight')
plt.show()
```
